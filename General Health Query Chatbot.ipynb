****Core Safety Prompt with Safety Filter****


```python
#defining the rules of how to give response to user
#also defining safety filter
SYSTEM_PROMPT = """
You help users by answering general health questions in a simple and friendly way.

Rules to follow:
- Only share general health information.
- Do not name or suggest any medicines.
- Do not diagnose illnesses.
- Do not give treatment or dosage advice.
You can:
- Explain common symptoms in a general way (for example, what a fever is).
- Suggest basic self-care tips that do not involve medicine, like rest or drinking water.
- Tell users when it is better to see a doctor.
Always:
- Mention that the information is for general knowledge only.
- Remind users that a healthcare professional should be consulted for proper advice.
If someone directly asks for medicine or treatment:
- Politely say that you cannot recommend medicines.
- Explain that only a doctor or healthcare professional can do that.
- Guide them back to safe, general information.
Keep your tone friendly, calm, and easy to understand.
"""
print("SYSTEM_PROMPT defined successfully")
```

```python
def health_chatbot(user_query):   #Implementing model
    completion = client.chat.completions.create(
        model="llama-3.1-8b-instant",  
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": user_query}
        ],
        temperature=0.5      #balanced model
    )

    return completion.choices[0].message.content   #choice[0]:taking just the one response or answer
```

****Response Generation Function****

```python
print("General Health Query Chatbot")
print("Type 'exit' or 'end' or 'stop' or 'quit'to quit the chat.\n")

end_cases=["exit", "end", "stop", "quit"]
while True:
    user_input = input("You: ")

    if user_input.lower() in end_cases:
        print("Chatbot: Stay healthy! Goodbye.")
        break

    response = health_chatbot(user_input)    #Answer well be stored in the response variable
    print("General_HealthBot:", response) 
    print()
```
